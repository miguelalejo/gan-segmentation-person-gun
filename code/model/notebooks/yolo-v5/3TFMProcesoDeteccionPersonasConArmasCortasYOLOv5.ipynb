# CreaciÃ³n Conjuntos de Datos

```
listPersonasArmadas = []
for fileIamgePath in glob.glob("./detect-humans-guns/exp/crops/gun/*.jpg"):
    listPersonasArmadas.append(os.path.basename(fileIamgePath))
print("Nro imagenes",len(listPersonasArmadas))
```

```
!rm -rf ./personas-armadas-detectadas
!mkdir personas-armadas-detectadas
```

```
!pwd
```

```
import shutil
srcPath = "./personas-detectadas"
destPath = "./personas-armadas-detectadas/"

for imageName in listPersonasArmadas:    
    fileSrc = os.path.join(srcPath,imageName)
    if os.path.isfile(fileSrc):
      shutil.copy(fileSrc,destPath)
    else:
      print(fileSrc)
    #shutil.copy(src, dst, *, follow_symlinks=True)

```
